```python

from enum import unique

from torch import le
from pinatrace_dataset import read_trace


trace = read_trace()

pcs = [t[0] for t in trace]

unique_pcs = list(set(pcs))
edges = {p: set() for p in unique_pcs}

for i in range(1, len(pcs)):
    edges[pcs[i-1]].add(pcs[i])

for pc in unique_pcs:
    print(pc, len(edges[pc]))
```
